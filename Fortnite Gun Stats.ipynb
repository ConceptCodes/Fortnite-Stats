```python
#imports
%matplotlib inline
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt

#get data
csv = pd.read_csv('Data/Weapons/weapons.csv')

#turn into object to play with later
data = pd.DataFrame(csv)

#dictionary for rarity do we can graph it
# to find i just used data['Rarity'].unique and turned the console log of that into a dictionary
rarity_dict = {'Common' : 0,'Uncommon' : 1,'Rare' : 2,'Epic' : 3,'Legendary' : 4}

#initialize and store x and y variables
DPS = data['DPS']
Rarity = []
for i in data['Rarity']:
    Rarity.append(rarity_dict[i])
r = ('Common','UnCommon','Rare','Epic','Legendary')
r_pos = np.arange(len(r))
    
#setup graph for visualization
fig, ax = plt.subplots()
plt.ylabel('Damage Per Second')
plt.xticks(r_pos, r)
plt.title('DPS by Rarity')
colors = ['gray', 'green', 'blue', 'purple', 'yellow']
for i,val in enumerate(DPS):
    ax.scatter(Rarity[i],DPS[i],color=colors[Rarity[i]], alpha=0.7, edgecolors='none')
ax.grid(True)
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['Common','Uncommon','Rare','Epic','Legendary']
c=0
uc=0
e=0
r=0
l=0
target_dps = np.mean(DPS)
for j in range(len(DPS)):
    if Rarity[j] == 0 and DPS[j] > target_dps:
        c+= 1
    elif Rarity[j] == 1 and DPS[j] > target_dps:
        uc+= 1
    elif Rarity[j] == 2 and DPS[j] > target_dps:
        r+= 1
    elif Rarity[j] == 3 and DPS[j] > target_dps:
        e+= 1
    elif Rarity[j] == 4 and DPS[j] > target_dps:
        l+= 1  
sizes = [c, uc, r, e, l]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#Amount of guns in each category
c=0
uc=0
e=0
r=0
l=0
for j in range(len(Rarity)):
    if Rarity[j] == 0:
        c+= 1
    elif Rarity[j] == 1:
        uc+= 1
    elif Rarity[j] == 2:
        r+= 1
    elif Rarity[j] == 3:
        e+= 1
    elif Rarity[j] == 4:
        l+= 1  
tallies = [c, uc, r, e, l]
r = ('Common','UnCommon','Rare','Epic','Legendary')
r_pos = np.arange(len(r))
plt.barh(r_pos, tallies, align='center', alpha=0.8)
plt.yticks(r_pos, r)
plt.xlabel('Amount')
plt.title('Amount of Guns by Rarity')
plt.show() 
```
